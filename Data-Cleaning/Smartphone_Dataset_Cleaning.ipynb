```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('mobiles.csv')
df.head(3)
```

#creating new columns

```python
df['brand'] = df['mobile_name'].str.split(' ').str.get(0)
df['model'] = df['mobile_name'].str.split(' ').str.get(1)
df.drop(columns = 'mobile_name', inplace=True)
df.head(2)
```

```python
df['price']=df['price'].str.replace('₹','').str.replace(',','')
df.head(1)
```

```python
df['processor'] = df['processor'].str.split(',').str.get(0)
df.head(2)
```

```python
df['ram'] = df['storage'].str.split().str.get(0)
df['ram'].unique()
```

```python
df['rom'] = df['storage'].str.split(',').str.get(1).str.split().str.get(0)
df['rom'].unique()
```

```python
df.drop(columns='storage', inplace=True)
```

```python
df.head(2)
```

```python
'''
whenever applying regex, syntax:- str.extract('(regex)')
'''
df['battery_capacity'] = df['battery'].str.extract('(\d+\smAh)')
df['battery_capacity'] = df['battery_capacity'].str.replace('mAh','')
df.head(1)
```

```python
df['charging_capacity'] = df['battery'].str.extract('(\d+W)')
df['charging_capacity']=df['charging_capacity'].str.replace('W','')
df.drop(columns = 'battery', inplace=True)
```

```python
df['display_size'] = df['display'].str.extract('(\d+\.\d+\s*inches)')
df['display_size'] = df['display_size'].str.replace('inches','')
df.head(1)
```

```python
df['display_hz'] = df['display'].str.extract('(\d+\s*Hz)')
df['display_hz'] = df['display_hz'].str.replace('Hz','')
df.drop(columns='display', inplace=True)
df.head(1)
```

```python
print(df['camera'][0])
```

```python
df['front_camera'] = df['camera'].str.extract('(\d+\sMP\sFront)')
df['front_camera'] = df['front_camera'].str.replace('MP Front', '')
```

```python
'''
approach 1:-
df['rear_camera'] = df['camera'].str.extract('(\d+\s*MP\s*\+\s*\d+\sMP\s*\+\s*\d+\sMP)')
the only problem is that it only considers those mobiles which have triple camera, as our regex is looking for 3 such MPs. but, in the
dataset, we have single, dual and triple camera phones. hence, considering every phone is important.
approach 2:-
now, there is no phone in this world which isn't having a camera. so, keeping the 1st MP is compulsory. so \d+\s*MP for the 1st MP.
for the rest MPs, they are optional in order to consider the single camera phone. only if other MPs exist, will they be considered. for
that purpose, we have (?:\s*\+\s*\d+\s*MP){0,2}.
\s*\+\s*\d+\s*MP ----> for other MPs
?: ----> if other MPs (above equation) exists, consider, else don't.
{0,2} ----> quantifier, either the equation won't exist, will exist once, or will exist twice.
'''
#df['rear_camera'] = df['camera'].str.extract('(\d+\s*MP\s*\+\s*\d+\sMP\s*\+\s*\d+\sMP)')
df['rear_camera'] = df['camera'].str.extract('(\d+\s*MP(?:\s*\+\s*\d+\s*MP){0,2})')
df.drop(columns='camera',inplace=True)
df.head(2)
```

```python
df['connectivity'] = df['connectivity'].str.replace(',','')
```

```python
'''
we won't just the G's.
if we want just 1 G, we do \dG.
but, if we want n such Gs, we again have to use ?:
so (?:\dG\s*) ---> 1 G, be it 3G, 4G, or 5G. NOTE:- all were space separated, and hence i added \s* at the end.
but, we want all the 3 -----> {1,3}
'''
df['connectivity'] = df['connectivity'].str.extract('((?:\dG\s*){1,3})')
df.head()
```

#outlier detection and handling

```python
df['price'].skew() #not normal, skew = 3.7
df['rating'].skew() #normal
df['specs_score'].skew() #normal
df['ram'].skew() #not normal, skew = 12
df['rom'].skew() #not normal, skew = 1.34
df['battery_capacity'].skew() #normal
df['charging_capacity'].skew() #not normal, skew = 1.86
df['display_size'].skew() #not normal, skew = -5.15
df['display_hz'].skew() #normal
```

#(i) handling non-normal distribution outliers:-
1. price
2. ram
3. rom
4. charging_capacity
5. display_size

```python
#1. price ----> skewed dist
'''
prices extreme but not invalid hence kept
'''
import seaborn as sns
import matplotlib.pyplot as plt
sns.kdeplot(df['price'], fill=True)
plt.title('price dist')
plt.xlabel('price')
plt.show()

sns.boxplot(df['price'])
plt.title('price boxplot before')
plt.show()

#outlier elimination:-
q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)
iqr = q3-q1
max_lim = q3 + 1.5*iqr
min_lim = q1 - 1.5*iqr

print("Upper limit",max_lim)
print("Lower limit",min_lim)
```

```python
#2. ram ---> skewed dist
'''
capping outliers as extreme values are impossible
'''
sns.kdeplot(df['ram'], fill=True)
plt.title('ram dist')
plt.xlabel('ram')
plt.show()
sns.boxplot(df['ram'])
plt.title('ram boxplot before')
plt.show()
#outlier elimination:-
q1 = df['ram'].quantile(0.25)
q3 = df['ram'].quantile(0.75)
iqr = q3-q1
upper = q3 + 1.5*iqr
lower = q1 - 1.5*iqr

print("Upper",upper)
print("Lower",lower)
```

```python
df['ram'] = df['ram'].clip(upper = upper)
df['ram'] = df['ram'].clip(lower = lower)
sns.boxplot(df['ram'])
plt.title('ram boxplot after')
plt.show()
```

```python
#3. rom ---> dist neither skewed nor normal
sns.kdeplot(df['rom'], fill=True)
plt.title('rom dist')
plt.xlabel('rom')
plt.show()
sns.violinplot(x=df['rom'])
plt.show()
max_l = df['rom'].quantile(0.99)
min_l = df['rom'].quantile(0.01)
print('max_l=', max_l)
print('min_l=', min_l)
df['rom'].describe()
```

```python
df['rom']= df['rom'].clip(upper=max_l)
df['rom']= df['rom'].clip(lower=min_l)
sns.violinplot(x=df['rom'])
plt.show()
df['rom'].describe()
```

```python
#4. charging_capacity ---> dist skewed
'''
outliers are valid, though extreme. hence keeping them, as it might represent charging capacity of premium expensive phones.
'''
sns.kdeplot(df['charging_capacity'], fill=True)
plt.title('charging_capacity dist')
plt.xlabel('charging_capacity')
plt.show()
sns.boxplot(df['charging_capacity'])
plt.title('charging_capacity boxplot before')
plt.show()
df['charging_capacity'].describe()
```

```python
#5. display_size ---> dist skewed
'''
outliers valid and hence keeping for analysis
'''
sns.kdeplot(df['display_size'], fill=True)
plt.title('display_size dist')
plt.xlabel('display_size')
plt.show()
sns.boxplot(df['display_size'])
plt.title('display_size boxplot before')
plt.show()
q1 = df['display_size'].quantile(0.25)
q3 = df['display_size'].quantile(0.75)
iqr = q3-q1
upper_lim = q3 + 1.5*iqr
lower_lim = q1 - 1.5*iqr
print("upper limit=",upper_lim)
print("lower limit=",lower_lim)
df['display_size'].describe()
```

# (ii) Handling normal distribution outliers

1. rating
2. specs_score
3. battery_capacity
4. display_hz



```python
'''
if a dist is normal, there will be very mild outliers only. the skewness is pretty low so no extreme ends.
only if the outliers are impossible (invalid), we trim. or else, we keep them. no need to cap.
'''
```

```python
#1. rating:-
up = df['rating'].mean() + 3*df['rating'].std()
low = df['rating'].mean() - 3*df['rating'].std()
print('upper limit=', up)
print('lower limit=', low)
```

```python
'''
no impossible values in dist. so keeping them
'''
df['rating'].describe()
```

```python
#2. specs_score
'''
keeping
'''
up = df['specs_score'].mean() + 3*df['specs_score'].std()
low = df['specs_score'].mean() - 3*df['specs_score'].std()
print('upper limit=', up)
print('lower limit=', low)
df['specs_score'].describe()
```

```python
#3. battery_capacity
'''
keeping them
'''
up = df['battery_capacity'].mean() + 3*df['battery_capacity'].std()
low = df['battery_capacity'].mean() - 3*df['battery_capacity'].std()
print('upper limit=', up)
print('lower limit=', low)
df['battery_capacity'].describe()
```

```python
#4. display_hz
'''
keeping
'''
up = df['display_hz'].mean() + 3*df['display_hz'].std()
low = df['display_hz'].mean() - 3*df['display_hz'].std()
print('upper limit=', up)
print('lower limit=', low)
df['display_hz'].describe()
```
